## **Problema #2**


Scrieti functii pentru a determina al n-lea numar fibonacci si pentru a determina cel mai mic numar fibonacci din intervalul [a,b]

```python
def fibo(n):
    n1 = 1
    n2 = 1
    if n == 1 or n == 2:
        return 1
    k = 2
    while k < n:
        n3 = n1 + n2
        n1 = n2
        n2 = n3
        k += 1
    return n3


def fibo2(a, b):
    n1 = 1
    n2 = 1
    if a <= n1 <= b:
        return n1
    n3 = 1
    while n3 < a:
        n3 = n1 + n2
        n1 = n2
        n2 = n3
    if n3 > b:
        return 'nu exista un numar fibo in [a,b]'
    return n3
```

## **Problema #3**

Interschimbati cifra minima cu cifra maxima a unui numar, folosind o functie.

```python
def swap_min_max(n):
    ls = []
    while n:
        ls.append(n % 10)
        n //= 10
    ls.reverse()
    cmin = min(ls)
    cmax = max(ls)
    poz1 = ls.index(cmin)
    poz2 = ls.index(cmax)
    ls[poz1], ls[poz2] = ls[poz2], ls[poz1]
    string_cifre = [str(cifra) for cifra in ls]
    m = int(''.join(string_cifre))
    return (int(m))


print(swap_min_max(123456789))
```

## **Problema #4**

Scrieti o functie care determina suma prefixelor pare ale unui numar n.
1234 are prefixele pare 12 si 1234 => 1

```python
def suma_prefixe_pare(n):
    ls = []
    suma = 0
    while n:
        ls.append(n % 10)
        n //= 10
    ls.reverse()

    for i in range(1, len(ls)):
        pref = 0
        for j in range(i + 1):
            pref = pref * 10 + ls[j]
        if pref % 2 == 0:
            suma += pref
    return suma


print(suma_prefixe_pare(1234))
```

## **Problema #5**

Scrieti o functie care returneaza produsul cifrelor diferite de k ale unui numar n.

```python
def prodk(n, k):
    p = 1
    while n:
        if n % 10 != k:
            p *= n % 10
        n //= 10
    return p


print(prodk(1234, 4))
```

```python
def nr_cifre_pare_impare(n):
    pare = []
    impare = []
    while n:
        if n % 2 == 1:
            impare.append(1)
        else:
            pare.append(1)
        n //= 10
    nr_pare = sum(pare)
    nr_impare = sum(impare)
    return nr_pare, nr_impare


print(nr_cifre_pare_impare(1234567890))
```

## **Problema #8**

Să se determine numărul din intervalul [1,n] care are număr maxim de divizori.
Dacă există mai multe asemenea numere, se va afișa cel mai mic dintre ele.

```python
def NrDivMax(n):
    def nrdiv(m):
        dm = 0
        for d in range(1, m + 1):
            if m % d == 0:
                dm += 1
        return dm

    ls = [0] * (n + 1)
    for j in range(1, n + 1):
        ls[j] = nrdiv(j)

    val_max = max(ls)
    poz = ls.index(val_max)
    return poz


print(NrDivMax(20))
```

## **Problema #9**

Scrieți un program care citește de la tastatură un text de tipul precizat și construiește în memorie, apoi afișează pe ecran, un șir de caractere ce reprezintă acronimul corespunzător numelui citit. Acronimul este format din primul caracter al fiecărui cuvânt al numelui care începe cu majusculă.

```python
def acronim(cuvinte):
    acro = []
    for cuvant in cuvinte:
        if 'A' <= cuvant[0] <= 'Z':
            acro += cuvant[0]
    return acro


ls = [x for x in input().split()]
string = acronim(ls)
string = ''.join(string)
print(string)
```

## **Problema #10**

Se se afișeze șirul citit astfel încât cuvintele de lungime maximă să fie înlocuite cu inversul (oglinditul) lor, restul cuvintelor și aranjarea lor în propoziție să rămână neschimbate.

```python
def inversmax(l):
    maxi = 0
    for i in range(len(l)):
        if len(l[i]) > maxi:
            maxi = len(l[i])
    for i in range(len(l)):
        if len(l[i]) == maxi:
            l[i] = ''.join(reversed(l[i]))
    return l


ls = [x for x in input().split()]
print(inversmax(ls))
```

## **Problema #11**

Scrieţi un program care citeşte de la tastatură un număr natural n şi construieşte în memorie o matrice cu n linii şi n coloane ale cărei elemente sunt numere naturale, fiecare reprezentând ultima cifră a câte unui termen al şirului lui Fibonacci, începând de la termenul de indice 1 şi până la termenul de indice n2 .

```python
def construieste_matrice_fibonacci(n):
    m = n * n
    ucfib = [1, 1]

    n1 = 1
    n2 = 1

    for _ in range(2, m):
        n3 = (n1 + n2) % 10
        ucfib.append(n3)

        n1 = n2
        n2 = n3

    matrice = []
    k = 0

    for i in range(n):
        linie = []
        for j in range(n):
            linie.append(ucfib[k])
            k += 1
        matrice.append(linie)

    return matrice


def afiseaza_matrice(matrice):
    for linie in matrice:
        print(*linie, sep='\t')


matricea_fib = construieste_matrice_fibonacci(4)
afiseaza_matrice(matricea_fib)
```

## **Problema #15**

Să se implementeze un program care primește ca intrare o propoziție și sortează cuvintele din aceasta în funcție de numărul de vocale din fiecare cuvânt. Dacă două cuvinte au același număr de vocale, acestea vor fi sortate lexicografic în ordine crescătoare.

```python
def nr_vocale(cuvant):
    nr_voc = 0
    for litera in cuvant:
        if litera in "aeiouAEIOU":
            nr_voc += 1
    return nr_voc


cuvinte = input().split()
cuvinte.sort(key=lambda cuvant: (nr_vocale(cuvant), cuvant))
print(cuvinte)
```

b) Scrieți o funcție cu 3 parametri nenuli de tip întreg a, b și c care să verifice dacă aceștia pot fi considerați ca fiind numere scrise în baza 2 sau nu, folosind apeluri utile ale funcției definite anterior.

De exemplu, pentru numerele 1001, 11 și 100 funcția trebuie să returneze valoarea True, iar pentru numerele 1001, 17 și 100 trebuie să returneze valoarea False.

(Indicatie – alipire(alipire(a, b, c)) = 1)

```python
def alipire(*args):
    rezultat_str = ""

    for n in args:
        cifra_max_n = max(str(n))
        rezultat_str += cifra_max_n

    if not rezultat_str:
        return 0

    return int(rezultat_str)
```
