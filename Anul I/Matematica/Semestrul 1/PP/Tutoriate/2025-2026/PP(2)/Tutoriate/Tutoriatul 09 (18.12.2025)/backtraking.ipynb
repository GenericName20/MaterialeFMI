Problema 1

Scrieti un program care sa afiseze toate permutarile unui numar n.

```python
def validare(x,k):
    for i in range(k):
        if x[i] == x[k]:
            return False
    return True

def back(x,n,k):
    if k == n:
        print(x)
    else:
        for i in range(1,n+1):
            x[k] = i
            if validare(x,k) == True:
                back(x,n,k+1)

x = [0]*3
back(x,3,0)
```

Problema 4

Sa se scrie un cod care afiseaza permutari in care numerele pare stau pe pozitiile impare.(permutari cu puncte fixe).

```python
def validare(x,k):
        if x[k]%2 == 0 and x[k]!=k+1:
            return False
        for i in range(k):
            if x[i] == x[k]:
                return False
        return True

def back(x,n,k):
    if k == n:
        print(*x)
    else:
        for i in range(1,n+1):
            x[k] = i
            if validare(x,k) == True:
                back(x,n,k+1)


n = int(input("n= "))
x = [0]*n
back(x,n,0,)
```

Problema 6

Sa se scrie un program care afiseaza toate permutarile elementelor unei multimi.

```python
def validare(x,k):
    for i in range(k):
        if x[i] == x[k]:
            return False
    return True
def back(x,n,k,ls):
    if k == n:
        print(x)
    else:
        for i in range(len(ls)):
            x[k] = ls[i]
            if validare(x,k) == True:
                back(x,n,k+1,ls)

ls = [4,7,3]
x = [0]*len(ls)
ls.sort()
back(x,len(ls),0,ls)
```
