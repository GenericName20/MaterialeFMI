## Problema 4

Se dÄƒ un dicÈ›ionar care reprezintÄƒ tabelul de frecvenÈ›Äƒ al unor caractere Ã®ntr-o frazÄƒ. SÄƒ se afiÈ™eze caracterele Ã®mpreunÄƒ cu frecvenÈ›a lor Ã®n ordine invers lexicograficÄƒ.


```python
fraza = "Maine dimineata voi prinde un caine roscat de un picior si rotind-ul ca pe o elice cu adevarat o sa zbor"
d = {}
for i in fraza:
  d[i] = d.get(i, 0) + 1
print(d) #asa obtinem un dictionar de frecventa, metoda prezentata in curs
print(sorted(d.items(), reverse = True))
```

## **Problema 2 a) din testul v1**


```python
import pprint

liceu = {profil : dict() for profil in ["matematica-informatica", "stiinte-ale-naturii", "stiinte-sociologice"]}

def main():
    with open("date1.txt", "r") as file:
        for line in file:
            line = line.strip().split()
            cnp = line[0]
            nume = line[1]
            prenume = line[2]
            clasa = line[3]
            media = float(line[4])
            profilul = line[5]
            tuplu_elev = (cnp, nume, prenume, clasa, media, profilul)
            if clasa not in liceu[profilul]:
                liceu[profilul][clasa] = [tuplu_elev]
            else:
                liceu[profilul][clasa].append(tuplu_elev)
        pprint.pprint(liceu)
```
