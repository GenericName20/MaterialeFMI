# Problema 1

Scrie o funcție care primește un șir de caractere (o propoziție) și returnează un dicționar în care cheile sunt cuvintele unice din propoziție, iar valorile sunt de câte ori apare fiecare cuvânt.

```python
def contor_frecventa_cuvinte(propozitie):
    cuvinte = propozitie.lower().split()
    frecventa = {}
    for cuvant in cuvinte:
        #frecventa[cuvant] = frecventa.get(cuvant, 0) + 1
        if cuvant in frecventa:
            frecventa[cuvant]+=1
        else:
            frecventa[cuvant] =1
    return frecventa

text = "Python este usor, si Python este distractiv"

rezultat = contor_frecventa_cuvinte(text)
print(rezultat)
```

# Problema 2

Ai un dicționar care stochează prețurile unor fructe. Scrie o funcție care primește numele unui fruct (cheia) și dicționarul, și returnează prețul acestuia. Dacă fructul nu este găsit, ar trebui să returneze un mesaj specific.

```python
def cauta_pret_fruct(nume_fruct, preturi_fructe):
    if nume_fruct in preturi_fructe:

        return preturi_fructe[nume_fruct]
    else:
        return f"Ne pare rău, '{nume_fruct}' nu este în stoc."

preturi = {
    "mar": 2.5,
    "banana": 1.8,
    "portocala": 3.0
}

pret_mar = cauta_pret_fruct("mar", preturi)
pret_para = cauta_pret_fruct("para", preturi)

print(pret_mar)
print(pret_para)
```

# Problema 3
Ai un dicționar care reprezintă inventarul unui magazin, unde cheia este numele produsului, iar valoarea este cantitatea.

Calculează și returnează suma totală a tuturor cantităților.

```python
def suma_inventar(inventar):
    suma_totala = 0
    for cantitate in inventar.values():
        suma_totala += cantitate
    return suma_totala

magazin = {
    "lapte": 15,
    "paine": 20,
    "oua": 30,
    "unt": 5
}

total_produse = suma_inventar(magazin)
print(total_produse)
```

# Problema 4

Scrie o funcție care primește un dicționar și returnează un dicționar nou în care cheile și valorile sunt inversate.

Presupune că toate valorile sunt unice pentru a nu avea chei duplicate în noul dicționar.

```python
def inverseaza_dictionar(original):
    dictionar_inversat = {}
    for cheie, valoare in original.items():
        dictionar_inversat[valoare] = cheie
    return dictionar_inversat

coduri_tari = {
    "Romania": "RO",
    "Germania": "DE",
    "Franta": "FR"
}

coduri_inversate = inverseaza_dictionar(coduri_tari)
print(coduri_inversate)
```

# Problema 6
Ai un dicționar de nume de studenți și scorurile lor.

 Sortează studenții în funcție de scor, de la cel mai mare la cel mai mic, și in ordine lexicografica într-o listă de perechi (nume, scor).

```python
def sorteaza_dupa_scor(scoruri):

    # Folosim .items() pentru a obține o listă de perechi (cheie, valoare).
    # Functia sorted() primeste un argument 'key' care este o functie lambda.
    # Expresia lambda ia o pereche 'item' (nume, scor) si returneaza item[1] (scorul).
    # 'reverse=True' sorteaza de la mare la mic.

    studenti_sortati = sorted(scoruri.items(), key=lambda item: (item[1], item[0]))
    return studenti_sortati

scoruri_studenti = {
    "Ana": 95,
    "Bogdan": 78,
    "Cristina": 99,
    "Dan": 85
}

rezultat = sorteaza_dupa_scor(scoruri_studenti)
print("Clasamentul studenților (Scor Descrescător):")
print(rezultat)
# Output: [('Cristina', 99), ('Ana', 95), ('Dan', 85), ('Bogdan', 78)]
```

# Problema 7
Ai un dicționar cu nume de orașe și populația lor.

Sortează aceste orașe (cheile dicționarului) în funcție de lungimea numelui lor, de la cel mai scurt la cel mai lung.

```python
def sorteaza_dupa_lungimea_numelui(orase):
    # Folosim .keys() pentru a obține o listă de chei (numele orașelor).
    # Functia lambda ia o cheie 'nume' și returnează len(nume) (lungimea numelui).

    orase_sortate = sorted(
        orase.keys(),
        key=lambda nume: len(nume)
    )
    return orase_sortate

populatie_orase = {
    "Cluj": 324,
    "Bucuresti": 1883,
    "Sibiu": 147,
    "Timisoara": 319
}

rezultat = sorteaza_dupa_lungimea_numelui(populatie_orase)
print("Orașe sortate după lungimea numelui (Scurt la Lung):")
print(rezultat)
# Output: ['Cluj', 'Sibiu', 'Dan', 'Bucuresti', 'Timisoara']
```
